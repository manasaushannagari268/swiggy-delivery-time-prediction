# handling missing values

```python
from sklearn.impute import SimpleImputer
```

```python
age_impute=SimpleImputer(strategy='median')
df[['age']]=age_impute.fit_transform(df[['age']])
```

```python
rating_impute=SimpleImputer(strategy='median')
df[['ratings']]=rating_impute.fit_transform(df[['ratings']])
```

```python
weather_impute=SimpleImputer(strategy="most_frequent")
df[["weather"]]=weather_impute.fit_transform(df[["weather"]])
```

```python
traffic_impute=SimpleImputer(strategy="most_frequent")
df[["traffic"]]=traffic_impute.fit_transform(df[["traffic"]])
```

```python
mul_del_impute=SimpleImputer(strategy="median")
df[["multiple_deliveries"]]=mul_del_impute.fit_transform(df[["multiple_deliveries"]])
```

```python
festival_impute=SimpleImputer(strategy="most_frequent")
df[["festival"]]=festival_impute.fit_transform(df[["festival"]])
```

```python
citytype_impute=SimpleImputer(strategy="most_frequent")
df[["city_type"]]=citytype_impute.fit_transform(df[["city_type"]])
```

```python
pick_time_impute=SimpleImputer(strategy="median")
df[["pickup_time_minutes"]]=pick_time_impute.fit_transform(df[["pickup_time_minutes"]])
```

```python
ord_time_impute=SimpleImputer(strategy="median")
df[["order_time_hour"]]=ord_time_impute.fit_transform(df[['order_time_hour']])
```

```python
distance_impute=SimpleImputer(strategy="median")
df[["distance"]]=distance_impute.fit_transform(df[['distance']])
```

```python
x=df.drop(columns="time_taken")
y=df["time_taken"]
```

```python
x.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=23)
```

# Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,root_mean_squared_error,r2_score
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print("mean_squared_error",mean_squared_error(y_test,y_pred))
print("root_mean_squared_error",root_mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
```

```python
knn=LinearRegression()
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print("mean_squared_error",mean_squared_error(y_test,y_pred))
print("root_mean_squared_error",root_mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```

# Decision TreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
print("mean_squared_error",mean_squared_error(y_test,y_pred))
print("root_mean_squared_error",root_mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```

# XGBoost

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
```

```python
xgb=XGBRegressor(n_estimators=100,
                          learning_rate=0.089,
                          max_depth=8,
                          Algorithm__subsample=0.95,
                           random_state=23)
xgb.fit(x_train,y_train)
y_pred=xgb.predict(x_test)
print("mean_squared_error",mean_squared_error(y_test,y_pred))
print("root_mean_squared_error",root_mean_squared_error(y_test,y_pred))
print("R2 score",r2_score(y_test,y_pred))
```
