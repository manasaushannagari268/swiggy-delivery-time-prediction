## Business Problem
Swiggy processes millions of food-delivery journeys across hundreds of cities every month. Accurate Estimated Time of Arrival (ETA) at order placement and during delivery is core to the customer experience — it affects conversion, cancellations, ratings, and retention. Today’s ETA accuracy suffers from many sources of uncertainty: highly variable traffic patterns, weather events, restaurant preparation variability, rider behaviour and skill, order batching/multiple deliveries, and city-specific operational constraints. Inaccurate ETAs cause:
Reduced customer trust and lower repeat orders.
Increased cancellations and refund costs.


Higher support volume and operational interventions.


Sub-optimal rider allocation and increased rider idle or overtime costs.


## Problem Statement
Swiggy’s objective is to predict the delivery time (minutes) per order at the moment of order placement (and update it dynamically), with business-grade accuracy and uncertainty estimates so the platform can
 (a) show reliable ETAs to customers
 (b) optimize rider allocation.

# Pipeline creation

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.compose import ColumnTransformer
```

```python
numerical_pipeline=Pipeline([("Numerical impution",SimpleImputer(strategy="median")),
                            ("Feature Scaling",StandardScaler())])
numerical_pipeline
```

```python
nominal_pipeline=Pipeline([("nominal imputation",SimpleImputer(strategy="most_frequent")),
                          ("Encoding",OneHotEncoder())])
nominal_pipeline
```

```python
ordinal_pipeline=Pipeline([("Ordinal imputation",SimpleImputer(strategy="most_frequent")),
                          ("Encoding",OrdinalEncoder(categories=order))])
ordinal_pipeline
```

```python
Data_preprocessing = ColumnTransformer([
    ('Nominal transformation', nominal_pipeline, nominal),
    ("Ordinal transformation", ordinal_pipeline, ordinal),
    ('Numerical Transformation', numerical_pipeline, numerical)
])

Data_preprocessing
```

```python
from sklearn.model_selection import train_test_split
```

# Bayesan Optimizer

```python
def Objective(trial):
    # creating  search space
    n_estimators=trial.suggest_int("n_estimators",100,150,200)
    learning_rate=trial.suggest_float("learning_rate",0.01,0.1)
    Algorithm__max_depth=trial.suggest_int("Algorithm__max_depth",6,8)
    Algorithm__subsample=trial.suggest_float("Algorithm__subsample",0.8,1.0)
    
    # Training
    xgb=XGBRegressor(n_estimators=n_estimators,learning_rate=learning_rate,max_depth=Algorithm__max_depth,subsample=Algorithm__subsample)

    trial_pipeline = Pipeline(steps=[
        ('Data Preprocessing', Data_preprocessing), 
        ('Algorithm', xgb)])
    
    #scoring
    score=cross_val_score(trial_pipeline,x,y,scoring="r2",cv=kfcv,n_jobs=-1).mean()
    return score
```

```python
import optuna
```

```python
study=optuna.create_study(direction="maximize")
```

```python
study.optimize(Objective,n_trials=10)
```
